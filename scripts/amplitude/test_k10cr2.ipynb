```python
from pylablib.devices import Thorlabs
Thorlabs.list_kinesis_devices()                       # find the SN
```

```python
rot = Thorlabs.KinesisMotor("55001180", scale="K10CR1")  # degrees in/out
print(rot.is_homed())
```

```python
rot.setup_limit_switch(hw_kind_cw='make_home', hw_kind_ccw='make_home')
rot.setup_homing(velocity=10)
rot.home(sync=True, force=True)
print(rot.is_homed())
```

```python
rot.move_by(40)
rot.wait_move()
```

```python
print(rot.setup_limit_switch(hw_kind_cw='make_home', hw_kind_ccw='make_home'))
```

```python
rot.home(sync=False, force=True); import time; time.sleep(1); print(rot.get_status())
```

```python
rot.move_to(50.0)              # absolute, degrees
rot.wait_move()
```

```python
rot.stop(); rot.move_to(0); rot.wait_move(); rot.set_position_reference(0); print(rot.get_position())
```

```python
rot.setup_velocity(max_velocity=10, acceleration=10)   # deg/s, deg/s²
print(rot.get_velocity_parameters())                   # confirm both non-zero
rot.home(sync=False, force=True)
import time; time.sleep(0.5)
print(rot.get_status())                                # want shaft turning now
```

```python
rot.get_position()
```

```python
print(rot.is_homed())                 # if True, that's why home() no-ops
print(rot.get_scale_units())          # must be 'deg' — confirms K10CR1 detected
rot.setup_homing(velocity=10)         # writes 0x0440; deg/s, scaled by 7.33e6
print(rot.get_homing_parameters())    # CONFIRM velocity now ~10, not 0.01
rot.home(sync=False, force=True)      # force past the homed bit; async so you can watch
import time; time.sleep(0.3)
print(rot.get_status())               # expect 'homing' in the list while it runs
rot.wait_for_home(timeout=60) 
```

```python
rot = Thorlabs.KinesisMotor("55001180", scale="K10CR1")  # degrees in/out

rot.home(sync=True)            # required for absolute angle after power-up
rot.move_to(45.0)              # absolute, degrees
rot.wait_move()
rot.move_by(-10.0)             # relative, degrees
rot.wait_move()
print(rot.get_position())      # degrees
```
